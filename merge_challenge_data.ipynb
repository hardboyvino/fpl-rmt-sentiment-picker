```python
import pandas as pd
import numpy as np
```

```python
df_1 = pd.read_csv('mod_transferable_players_joy.csv')
df_2 = pd.read_csv('mod_transferable_players_tosin.csv')
df_3 = pd.read_csv('mod_transferable_players_niyi.csv')

df_1.shape, df_2.shape, df_3.shape
```

```python
df_1.head()
```

```python
df_2.head()
```

```python
df_3.head()
```

```python
merged_df_1_2_3 = df_1.merge(
                df_2,
                how='inner',
                on=['Name', 'Team', 'Price', 'Position']
            ).merge(
                df_3,
                how='inner',
                on=['Name', 'Team', 'Price', 'Position']
            )
```

```python
merged_df_1_2_3.head()
```

```python
merged_df_1_2_3['Points'] = (merged_df_1_2_3['Points_x'] + \
                                merged_df_1_2_3['Points_y'] + \
                                merged_df_1_2_3['Points']) / 3
```

```python
merged_df_1_2_3.sort_values(by='Points', ascending=False, inplace=True)
```

```python
merged_df_1_2_3.head()
```

```python
merged_df_1_2_3[['Name', 'Price', 'Position', 'Team', 'Points']].to_csv('transferable_combined.csv', index=False)
```
